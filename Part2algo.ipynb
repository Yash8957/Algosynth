```python

# Part 2: Multiple Number Checker

# Initialize an empty list to store numbers
numbers = []

# Collect 5 numbers from the user
count = 0
while count < 5:
    num = int(input(f"Enter number {count + 1}: "))
    numbers.append(num)
    count += 1

# Check parity for each number
for num in numbers:
    if num % 2 == 0:
        print(f"{num} is Even")
    else:
        print(f"{num} is Odd")
```
